```python
# Carregando as bibliotecas
import pandas as pd
from pandas import read_csv
from pandas.plotting import scatter_matrix
from matplotlib import pyplot
```

```python
# Carregando o Dataeet
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
names = ['class','alcohol', 'malic-acid', \
         'ash', 'alcalinity', 'magnesium', \
         'phenols','flavanoids','nonflavanoid', \
         'proanthocyanins','color-intensity','hue',\
         'od280','proline']
wine = read_csv(url, names=names)
print(wine.head(5))
```

```python
# Conhecendo o Dataeet
print("Descrição do Dataset"+"\n")
print(wine.describe())


print("\n\n\n"+"Categorias\n")
print(wine.groupby('class').size())

print("\n\n\n"+"BoxPlot\n")
wine.plot(kind='box', subplots=True, layout=(2,7), sharex=False, sharey=False)
pyplot.show()
wine.plot(kind='box')
pyplot.show()
```

```python
# DataSet Hot Encoding
wine_encoded = pd.get_dummies(wine['class'])
wine_encoded = pd.concat([wine[['alcohol', 'malic-acid', 'ash', 'alcalinity', \
                                'magnesium','phenols','flavanoids',\
                                'nonflavanoid','proanthocyanins',\
                                'color-intensity','hue','od280','proline']],\
                          wine_encoded],axis=1)
wine_encoded
```
